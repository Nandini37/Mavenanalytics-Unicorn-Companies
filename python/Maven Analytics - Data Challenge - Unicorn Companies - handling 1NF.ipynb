# Maven Data Challenge

Cleaning Unicorn Data Challenge.

- Casting text to numeric
- Exploding rows

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Unicorn_Companies.csv")
```

```python
print("Shape:" ,df.shape)

df.dropna(inplace=True)

print("Dataframe Shape after drop:" ,df.shape)
```

```python
df['Valuation_int'] = df['Valuation']
df['Funding_int'] = df['Funding']

df['Funding_int'].value_counts()
df = df[df['Funding'] != 'Unknown']
```

```python
df['Valuation_int'] = df['Valuation_int'].str.replace('$','').str.replace('B','000000000').str.replace('M','000000')
df['Funding_int'] = df['Funding_int'].str.replace('$','').str.replace('B','000000000').str.replace('M','000000')

df['Valuation_int'] = pd.to_numeric(df['Valuation_int'])

df['Funding_int'] = pd.to_numeric(df['Funding_int'])


df.head()
```

```python
df = df.reset_index()
df.head()
```

```python
df['Select Investors'] = df['Select Investors'].str.split(",")
```

```python
df = df.explode("Select Investors")
```

```python
df.isnull().sum()*100/df.shape[0]
```

```python
#df[df["Company"]== "SVOLT"]
df[df["Company"]== "Evidation"]
df = df.replace(r'^s*$', float('NaN'), regex = True)
df = df.dropna() 
```

```python
df[df["Company"]== "Evidation"]
df[df["Company"]== "SVOLT"]
```

```python
df.to_csv("Unicorn_Companies_clean.csv")
del df
```
